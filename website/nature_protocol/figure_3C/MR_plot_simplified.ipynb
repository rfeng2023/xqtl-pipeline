```r
library("tidyverse")
```

```r
data = readRDS("figure_3C.rds")
```

```r
plot_df = data[[1]]
plot_df_2 = data[[2]]
```

```r

plot_df %>%
  mutate(
    cs = as.factor(cs_coverage_0.7)
  ) %>%
  ggplot(aes(x = start, y = z, color = cs)) +
  geom_hline(yintercept = 0, linewidth = 0.4) +
  geom_segment(aes(xend = start, y = 0, yend = z),
               linewidth = 0.4) +
  geom_point(size = 2) +
  # tweak colors as needed
  scale_color_manual(values = c("0" = "grey70", "1" = "blue")) +
  labs(
    x = "Chromosome 5 position",
    y = "Z"
  ) +
  theme_classic(base_size = 14) +
  theme(legend.position = "none")+facet_grid(event_ID~.,scales = "free")
```

```r


plot_df_2 %>%
  mutate() %>%
  ggplot(aes(x = z_dlpfc, y = z_ad)) +
  geom_point(size = 2) +
  scale_color_manual(values = c("0" = "grey70", "1" = "blue")) +
  labs(x = "DLPFC Z", y = "AD Z") +
  theme_classic(base_size = 14) +geom_abline(slope = 0.004, intercept = 0, linewidth = 0.6)+annotate("text", x = Inf, y = -Inf, hjust = 1.1, vjust = -0.5,
           label = "β = 0.004\nSE = 0.001\nP = 0.001\nQ pval = 0")
```
